```python
import pandas as pd
```

```python
x = pd.read_csv("sti_rates.csv")

x = x.fillna(0)

x = x.replace('<5', 0)

x['Reported Cases'] = x['Reported Cases'].astype(int)

x.columns
```

```python
lg_df = x[(x['Gender'] == 'MALE') & (x['STI Name'] == 'GON')]
```

```python


pd.DataFrame(lg_df.groupby('Year')['Reported Cases'].sum())

```

```python
x = x.groupby(['Community Area Name', 'Year'])['Reported Cases'].sum().reset_index()

x = x[x['Year'] == 2014]
x
```

```python
m_chl = x[(x['Gender'] == 'MALE') & (x['STI Name'] == 'CHL')]
m_chl = m_chl.groupby('Year')['Reported Cases'].sum()
m_chl
```

```python
f_chl = x[(x['Gender'] == 'FEMALE') & (x['STI Name'] == 'CHL')]
f_chl = pd.DataFrame(f_chl.groupby('Year')['Reported Cases'].sum())
f_chl
```
