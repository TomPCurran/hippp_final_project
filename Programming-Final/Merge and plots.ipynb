```python
import pandas as pd 
```

```python
%matplotlib inline
```

```python
con_dist = pd.read_csv("condom_dist_u.csv", index_col = "Community Area Name")
#con_dist.head(5)
fgon = pd.read_csv("gonorrhea_females.csv", index_col = "Community Area Name")
mgon = pd.read_csv("gonorrhea_males.csv", index_col = "Community Area Name")
fchla = pd.read_csv("chlamydia_females.csv", index_col = "Community Area Name")
mchla = pd.read_csv("chlamydia_males.csv", index_col = "Community Area Name")
#fgon.head(5)
```

```python
con_dist_group = con_dist.groupby(["Community Area Name"]).count()
con_dist_group.dropna(inplace = True)
con_dist_group.rename(columns = {"Name" : "Number of Condom Centers"}, inplace = True)
con_dist_group.head(5)
con_fgon_merged = con_dist_group.join(fgon)
con_fgon_merged
```

```python
graph1 = con_fgon_merged.plot(kind = "scatter", x = "Number of Condom Centers", y = "Incidence Rate 2000")
graph1.set_ylabel("Female Gonorrhea Incidence Rate 2000")

```

```python
graph2 = con_fgon_merged.plot(kind = "scatter", x = "Number of Condom Centers", y = "Incidence Rate 2014")
graph2.set_ylabel("Female Gonorrhea Incidence Rate 2014")
```
